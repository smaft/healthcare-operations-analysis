### Key Insights
- Service duration is highly right-skewed, with a small percentage of encounters
  consuming disproportionate time.
- Clear peak demand hours are observed, indicating temporal congestion patterns.
- Variability in service duration suggests inconsistent operational processes.


### Recommendations
- Align staffing levels with identified peak hours to reduce congestion.
- Standardize workflows for high-variability services to reduce delays.
- Track service duration KPIs continuously to detect emerging bottlenecks.


### Limitations
This analysis is based on synthetic or simplified healthcare data and may not
capture the full complexity of real-world hospital operations.
