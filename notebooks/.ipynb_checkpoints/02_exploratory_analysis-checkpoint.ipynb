```python
## This notebook explores service duration patterns and temporal trends to identify operational bottlenecks.
```

```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
sns.set_style("whitegrid")
```

```python
df = pd.read_csv("../data/processed/encounters_cleaned.csv", parse_dates=["START", "STOP"])
```

```python
print(f"Data Loaded: {df.shape[0]} records")
df.head()
```

```python
df["service_duration_min"].describe()
```

```python
plt.figure(figsize=(12, 6))
```

```python
sns.histplot(df["service_duration_min"], bins=50, kde=True, color='skyblue', edgecolor='black')
```

```python
plt.title("Distribution of Service Duration (Patient Flow)", fontsize=16)
plt.xlabel("Duration in Minutes", fontsize=12)
plt.ylabel("Frequency (Number of Patients)", fontsize=12)
```

```python
plt.axvline(df["service_duration_min"].mean(), color='red', linestyle='--', label=f'Mean: {df["service_duration_min"].mean():.1f} min')
plt.axvline(df["service_duration_min"].median(), color='green', linestyle='-', label=f'Median: {df["service_duration_min"].median():.1f} min')
plt.legend()
plt.show()
```

```python
df["arrival_hour"] = df["START"].dt.hour

print("Busy Hours (Top 5):")
print(df["arrival_hour"].value_counts().head())
```

```python
hourly_counts = df["arrival_hour"].value_counts().sort_index()
```

```python
plt.figure(figsize=(12, 6))
hourly_counts.plot(kind='bar', color='coral', width=0.8)
```

```python
plt.title("Peak Arrival Hours (Patient Congestion)", fontsize=16)
plt.xlabel("Hour of Day (0-23)", fontsize=12)
plt.ylabel("Number of Arrivals", fontsize=12)
plt.xticks(rotation=0)
plt.grid(axis='y', alpha=0.5)

plt.show()
```

```python

dept_col = 'DESCRIPTION' if 'DESCRIPTION' in df.columns else 'DEPARTMENT'

if dept_col in df.columns:
   
    dept_duration = df.groupby(dept_col)["service_duration_min"].mean().sort_values(ascending=True).tail(10) 

    plt.figure(figsize=(10, 8))
    dept_duration.plot(kind='barh', color='teal')
    
    plt.title(f"Top 10 Services by Average Duration", fontsize=16)
    plt.xlabel("Average Duration (Minutes)", fontsize=12)
    plt.ylabel("Service / Department", fontsize=12)
    
    plt.show()
else:
    print(f"Column '{dept_col}' not found. Please check column names.")
```

##  Summary of Findings
1. **Distribution:** The service duration is **right-skewed**; most visits are short, but outliers significantly increase the average.
2. **Peak Hours:** There is a clear pattern in arrival times (check the coral chart above). Staffing should be increased during these peak hours.
3. **Bottlenecks:** Certain services (shown in the teal chart) have much higher average durations, making them prime targets for process optimization.
