```python
## Initial data quality checks were performed to assess missing values,duplicates, and variable distributions.
```

```python
df["START"] = pd.to_datetime(df["START"])
df["STOP"] = pd.to_datetime(df["STOP"])
```

```python
df["service_duration_min"] = (df["STOP"] - df["START"]).dt.total_seconds() / 60
```

```python
df = df[df["service_duration_min"] > 0] 
```

```python
df.to_csv("../data/processed/encounters_cleaned.csv", index=False)
```
