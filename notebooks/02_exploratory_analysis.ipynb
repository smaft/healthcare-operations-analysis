```python
## This notebook explores service duration patterns and temporal trends to identify operational bottlenecks.
```

```python
import pandas as pd
```

```python
df = pd.read_csv("../data/processed/encounters_cleaned.csv")
```

```python
df["service_duration_min"].describe()
```

```python
import matplotlib.pyplot as plt
```

```python
df["service_duration_min"].hist(bins=40)
plt.xlabel("Service Duration (min)")
plt.ylabel("Count")
plt.show()
```

```python
df["hour"] = pd.to_datetime(df["START"]).dt.hour
df["hour"].value_counts().sort_index()
```

```python
if "DEPARTMENT" in df.columns:
    df.groupby("DEPARTMENT")["service_duration_min"].mean().sort_values(ascending=False)
```

## Service duration shows right-skewed distribution with clear peak hours, indicating demand concentration during specific periods.
